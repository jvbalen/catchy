# Catchy feature extraction

Extracting catchy features from a folder of files will involve three steps:

#### 1. Base feature extraction
Currently MFCC, chroma, melody and perceptual features but really anything you want.
This part will heavily rely on external code, but should also be the easiest to work around.

#### 2. Pitch descriptor extraction
Compute mid-level pitch descriptors from chroma and/or melody 
information computed in step one.
Essentially an implementation of several kinds of audio bigram descriptors.

#### 3. Feature transforms
Compute first and second order aggregates of any of the features computed in step 1 and step 2.

#### How to use

Each of these steps has its own python file that can be imported to run it on some dataset.
